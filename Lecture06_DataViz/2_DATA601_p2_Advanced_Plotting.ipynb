```python
import numpy as np
import matplotlib.pyplot as plt

# Cluster data
a = np.random.uniform(0, 2*np.pi, size=10000)
r = np.random.exponential(size=10000)

# Make a polar plot.
plt.polar(a, r, linestyle='none', marker='o')
plt.show()
```

```python
r = np.random.exponential(size=10000)
plt.plot(r)
```

```python
# Alpha controls transparency
# 0 = transparent
# 1 = opaque

# Make markers transparent.
plt.clf()
plt.polar(a, r, linestyle='none', marker='o', alpha=0.2)
plt.show()
```

```python
# Imshow is the go-to image plotting function in matplotlib.

# Small parabolic image: z=x*(y-0.5)
x = np.linspace(0, 1, 21)
y = np.linspace(0, 1, 21) - 0.5
z = np.outer(x, y)

# Show it.
plt.clf()
plt.imshow(z)

# But this is probably not quite what you want...
```

```python
# Manually convert to Cartesian
plt.clf()
plt.imshow(z[:,::-1].T)

# Or do it the easy way...
plt.clf()
plt.imshow(z.T, origin='lower')
```

```python
# Smooth with bilinear
# interpolation.
plt.clf()
plt.imshow(z.T, origin='lower', interpolation='bilinear')
```

```python
# Pixels centered on pixel number. Change the ranges to stretch.
plt.clf()
plt.imshow(z.T, origin='lower', extent=[0, 5, 0, 10])
```

```python
# Change pixel aspect ratio to match.
# axes aspect ratio (default is 1:1).
plt.clf()
plt.imshow(z.T, origin='lower', aspect='auto')
```

```python
# imshow autoscales to color map. Change range of data mapped.
plt.clf()
plt.imshow(z.T, origin='lower', vmin=-0.3, vmax=0.6)
```

```python
# Change the color map.
plt.clf()
plt.imshow(z.T, origin='lower', cmap=plt.cm.gray)
```

```python
# Default colorbar is simple. 
plt.clf()
plt.imshow(z.T, origin='lower')
plt.colorbar()
```

```python
# Change the colorbar orientation.
plt.clf()
plt.imshow(z.T, origin='lower')
plt.colorbar(orientation='horizontal')
```

```python
# PCOLOR: Useful for arbitrary orientations and pixel sizes.
# Make new x and y values for pixel edges.
xa = np.linspace(-1, 0, 21)
ya = np.linspace(-2, -1, 21)
# Replot with these axes.
plt.clf()
plt.pcolor(xa, ya, z)
```

```python
# contour plots
# Make new data.
x = np.linspace(-1, 1, 101)
y = np.linspace(-1, 1, 101)
z = np.sin(np.outer(x, y))
# Contour plot the new data.
plt.clf()
plt.contour(z)
```

```python
# Specify the number of contour lines.
plt.clf()
plt.contour(z, 20)
```

```python
# Add contour labels.
plt.clf()
c1 = plt.contour(z)
plt.clabel(c1)
```

```python
# Overlay contours on images.
plt.clf()
plt.imshow(z)
plt.contour(z, colors='w')
```

```python
# HISTOGRAMs
# Gaussian random data
y = np.random.normal(size=10000)

# Default histogram (10 bins)
plt.clf()
plt.hist(y)
```

```python
# Change the number of bins.
plt.clf()
plt.hist(y, bins=20)
```

```python
# Change the bin edges.
plt.clf()
edges=[-3, -2.5, -2, -1, -0.5, 0, 1, 1.5, 2, 2.5]
plt.hist(y, bins=edges)
```

```python
# Just show the lines.
plt.clf()
plt.hist(y, histtype='step')
```

```python
# 3x1 grid of plots.
plt.clf()
x = np.linspace(-1, 1, 100)
y1 = x
y2 = x**2
y3 = x**3

# Make the plots one at a time.
ax1 = plt.subplot(3, 1, 1)
ax1.plot(x, y1)
ax2 = plt.subplot(3, 1, 2)
ax2.plot(x, y2)
ax3 = plt.subplot(3, 1, 3)
ax3.plot(x, y3)
```

```python
# Arbitrary subplot placement.
plt.clf()

# Make the plots one at a time.
ax1 = plt.axes([0.1, 0.1, 0.9, 0.8])
ax1.plot(x, y1)
ax2 = plt.axes([0.65, 0.25, 0.2, 0.2])
ax2.plot(x, y2)
ax3 = plt.axes([0.20, 0.55, 0.2, 0.2])
ax3.plot(x, y3)
```

```python
#  TEXT ANOTATIONS
# Just specify location and text. Location in axis units!
plt.clf()
plt.plot(x, y3)
plt.text(-0.25, 0.15, 'This is sdfsar.')
```

```python
# EQUAL AXES 
import matplotlib.pyplot as plt
import numpy as np

# Let's create a circle of radius 3.
an = np.linspace(0, 2 * np.pi, 100)
x = 3 * np.cos(an)
y = 3 * np.sin(an)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y)
axs[0, 0].set_title('not equal, looks like ellipse', fontsize=10)

axs[0, 1].plot(x, y)
axs[0, 1].axis('equal')
axs[0, 1].set_title('equal, looks like circle', fontsize=10)
```

```python
# EQUAL AXES 
import matplotlib.pyplot as plt
import numpy as np

# Let's create a circle of radius 3.
an = np.linspace(0, 2 * np.pi, 100)
x = 3 * np.cos(an)
y = 3 * np.sin(an)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y)
axs[0, 0].set_title('not equal, looks like ellipse', fontsize=10)

axs[0, 1].plot(x, y)
axs[0, 1].axis('equal')
axs[0, 1].set_title('equal, looks like circle', fontsize=10)

axs[1, 0].plot(x, y)
axs[1, 0].axis('equal')
axs[1, 0].axis([-3, 3, -3, 3])
axs[1, 0].set_title('still a circle, even after changing limits', fontsize=10)

axs[1, 1].plot(x, y)
axs[1, 1].set_aspect('equal', 'box')
axs[1, 1].set_title('still a circle, auto-adjusted data limits', fontsize=10)

fig.tight_layout()

plt.show()
```

```python
# EQUAL AXES 
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
f1 = np.cos(x)
f2 = np.sin(x)

fig, axs = plt.subplots(1, 2)

axs[0].plot(x, f1)
axs[0].set_xlabel('$\\theta$ (Deg)')
axs[0].set_ylabel('Cos($\\theta$)')

axs[1].plot(x, f2, color='r')
axs[1].set_xlabel('$\\theta$ (Deg)')
axs[1].set_ylabel('Sin($\\theta$)')

fig.tight_layout()

plt.show()
```
